```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
def plotting(x):
    df = pd.read_excel(x)
    df.set_index("Muestras", inplace=True)
    media = df.mean()
    var = df.var()
    log_media=np.log(media)
    log_var=np.log(var)
    x = log_media.values.reshape(-1,1)
    y = log_var.values.reshape(-1,1)   
    model = LinearRegression()
    model.fit(x, y)
    r_sq = model.score(x, y)
    print('coefficient of determination:', r_sq)
    print('intercept:', model.intercept_)   #la n
    print('slope:', model.coef_)            #la m
    print("La ecuaci贸n de la recta es la siguiente: f(x)=", model.coef_, "x +",model.intercept_)
    
    x_mean, y_var = pd.Series(log_media, name="log_media"), pd.Series(log_var, name="log_var")
    sns.regplot(x_mean , y_var, data=df, scatter=True, fit_reg=True, ci=None, marker = "+")

```

```python
plotting("118ALLL.xlsx")
print("\nResultados de todos los datos")
```

```python
plotting("20ALLL.xlsx")
print("\nResultados de todos los datos parseados por 20")
```
