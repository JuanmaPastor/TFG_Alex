```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from scipy.stats import norm
from scipy import stats
```

```python
def plotting(x):
    df = pd.read_excel(x)
    df.set_index("Muestras", inplace=True)
    media = df.mean()
    log_media=np.log(media)
    med=media.values.tolist()

    X_scaled = preprocessing.scale(log_media)
    sns.distplot(X_scaled, kde=True, hist = True, fit = norm)
    
    plt.title('Distribuci贸n de las medias relativas de las bacterias')
    plt.xlabel('Log de la media estandarizado')
    plt.ylabel('Densidad')
    print(stats.shapiro(X_scaled)) #El primero es el Test de Saphiro Wilk y el segundo el p value
    
    

```

```python
plotting("None0_Affected_Lung.xlsx")
print("\nResultados de los datos de pulm贸n afectado en todas las bacterias")
```
